## ETS Decomposition

```python
# Question No. 1
# Using ETS Decomposition, displaying the trend graph alone (classified_ts_01.csv), What is the type of trend?
# Support your answer with graph size (30,5)
ts01 = seasonal_decompose(classified_1['Value'], model='mul')
ts01.trend.plot(figsize = (30,5), grid=True)
```

```python
# Question No. 2
# Using ETS Decomposition, displaying the seasonality alone (classified_ts_01.csv), What can you tell about the presence of seasonality?
# Support your answer with graph size (30,5)

seasonal_3yr = ts01.seasonal.loc['1980-01-31':'1983-01-31']
seasonal_3yr.plot(figsize=(30, 5), grid=True, marker="o")
```

```python
# Question No. 3
# Using ETS Decomposition, displaying the trend residuals alone (classified_ts_01.csv), What can you tell about the residuals?
# Support your answer with graph size (30,5)
ts01 = seasonal_decompose(classified_1['Value'], model='mul')
ts01.resid.plot(figsize = (30,5), grid=True)
```

```python
# Question No. 4
# Using ETS Decomposition, displaying the seasonality alone (classified_ts_02.csv), If there is seasonality, how many periods, if there is any?
# Support your answer with graph size (30,5)

ts02 = seasonal_decompose(classified_2['Value'], model='multiplicative', period=12)

# Extract the seasonal component for the 3-year range
seasonal_3year = ts02.seasonal.loc['1980-01-31':'1983-01-31']

# Corrected plotting line
seasonal_3year.plot(figsize=(30, 5), grid=True, marker="o")

# Optional: Add labels for better readability
plt.title("Seasonal Component (1980-1983)")
plt.xlabel("Date")
plt.ylabel("Seasonal Effect")
plt.show()
```

## Comparative Graph

```python
# Question No. 12
# Generate Graphs below woht at least 90% simililarity ('On Notebook')
# First Time Series Decomposition
decomp_ets = seasonal_decompose(classified_1["Value"], model="multiplicative", period=12)
stl = STL(classified_1["Value"], period=12, robust=True)
decomp_stl = stl.fit()

# Second Time Series Decomposition
decomp_ets_2 = seasonal_decompose(classified_2["Value"], model="multiplicative", period=12)
stl_2 = STL(classified_2["Value"], period=12, robust=True)
decomp_stl_2 = stl_2.fit()

# Plot ETS and STL Decomposition Side-by-Side
fig, axes = plt.subplots(4, 4, figsize=(25, 20))

# Titles for the decomposition components
titles = ["Original Series", "Trend", "Seasonality", "Residuals"]

fig.suptitle("Comparative Graphs ETS vs STL Decomposition", )

# First Time Series
components_ets = [classified_1["Value"], decomp_ets.trend, decomp_ets.seasonal, decomp_ets.resid]
components_stl = [classified_1["Value"], decomp_stl.trend, decomp_stl.seasonal, decomp_stl.resid]

# Plot first series ETS (first column) and STL (second column)
for i, (comp_ets, comp_stl) in enumerate(zip(components_ets, components_stl)):
    comp_ets.plot(ax=axes[i, 0], title=f"ETS {titles[i]} (TS1)")
    axes[i, 0].grid()

    comp_stl.plot(ax=axes[i, 1], title=f"STL {titles[i]} (TS1)")
    axes[i, 1].grid()

#Second Time Series
components_ets_2 = [classified_2["Value"], decomp_ets_2.trend, decomp_ets_2.seasonal, decomp_ets_2.resid]
components_stl_2 = [classified_2["Value"], decomp_stl_2.trend, decomp_stl_2.seasonal, decomp_stl_2.resid]

# Plot second series ETS (third column) and STL (fourth column)
for i, (comp_ets_2, comp_stl_2) in enumerate(zip(components_ets_2, components_stl_2)):
    comp_ets_2.plot(ax=axes[i, 2], title=f"ETS {titles[i]} (TS2)")
    axes[i, 2].grid()

    comp_stl_2.plot(ax=axes[i, 3], title=f"STL {titles[i]} (TS2)")
    axes[i, 3].grid()


plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```
