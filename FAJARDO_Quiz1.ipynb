#Data Manipulation with Pandas

*   Handle Mising data
*   Group By Requirements



```python
import numpy as np
import pandas as pd

#Load csv file
df = pd.read_csv('https://gist.githubusercontent.com/kevin336/acbb2271e66c10a5b73aacf82ca82784/raw/e38afe62e088394d61ed30884dd50a6826eee0a8/employees.csv')
df.head()
```

```python
# Convert 'HIRE_DATE' to datetime objects
df['HIRE_DATE'] = pd.to_datetime(df['HIRE_DATE'])

# Display the updated DataFrame
df.head()
```

```python
#Handling missing data
df = df.drop(columns=['COMMISSION_PCT'])
df.head()
```

```python
# Arrange by employee id and reset the index
df = df.sort_values(by='EMPLOYEE_ID')
df = df.reset_index(drop=True)

# Convert 'MANAGER_ID' to numeric, handling errors
df['MANAGER_ID'] = pd.to_numeric(df['MANAGER_ID'], errors='coerce')

# Fill NaN values with the mean of the numeric 'MANAGER_ID' column
df['MANAGER_ID'] = df['MANAGER_ID'].fillna(df['MANAGER_ID'].mean())

# Remove decimals from 'MANAGER_ID'
df['MANAGER_ID'] = df['MANAGER_ID'].astype(int)

df.head()
```

```python
#salary average
S_average = df['SALARY'].mean()
print("Average Salary: ",S_average)

#Highest Salary
S_max= df['SALARY'].max()
print("Highest Salary: ",S_max)

#Lowest Salary
S_min= df['SALARY'].min()
print("Minimum Salary: ",S_min)
```

```python
# Group by 'DEPARTMENT_ID' and calculate the mean salary for each department
avg_salary_by_dept = df.groupby('DEPARTMENT_ID')['SALARY'].mean().sort_values(ascending=False)

# Display the result
avg_salary_by_dept
```

```python
# Group by department and count employees
employee_count_by_dept = df.groupby('DEPARTMENT_ID')['EMPLOYEE_ID'].count()

# Sort in descending order
employee_count_by_dept = employee_count_by_dept.sort_values(ascending=False)

# Display the result
employee_count_by_dept
```
