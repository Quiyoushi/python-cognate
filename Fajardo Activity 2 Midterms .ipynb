## ACTIVITY 2: Creating and Customizing Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt
x = [5,7,8,5,6,7,9,10,12,11]
y = [99,86,87,88,100,86,103,87,94,78]


plt.scatter(x, y)
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Scatter Plot Title')
plt.show()
```

```python
# Generate temperature data
temp_range = np.linspace(20, 38, 100)
temperature = temp_range + np.random.normal(0, 3, 100)

# Generate sales data
sales = 50 + 7 * temperature + np.random.normal(0, 20, 100)

# Create a categorical variable for ice cream flavors
flavors = np.random.choice(['Chocolate', 'Vanilla'], size=100)

# Create a color map for the flavors
colors = {'Chocolate': 'brown', 'Vanilla': 'grey'}

# Create the scatter plot with different colors for each flavor
for flavor in set(flavors):
    alpha_value = 0.4 if flavor == 'Vanilla' else 1  # Set alpha for Vanilla only
    plt.scatter(temperature[flavors == flavor], sales[flavors == flavor],
                color=colors[flavor], label=flavor, alpha=alpha_value)

plt.xlabel('Average Daily Temperature (Celsius)')
plt.ylabel('Ice Cream Sales (Units)')
plt.title('Ice Cream Sales vs. Temperature by Flavor')
plt.legend()
plt.show()
```
